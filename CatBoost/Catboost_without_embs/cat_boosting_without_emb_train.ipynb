```python
from sklearn.feature_extraction.text import CountVectorizer

def create_top_bag_of_words(sentences, top_words):
    vectorizer = CountVectorizer(max_features=top_words)
    bag_of_words = vectorizer.fit_transform(sentences)
    len_sentences = [len(sentence.split()) for sentence in sentences]

    return bag_of_words, len_sentences, vectorizer

def encode_sentences(sentences, vectorizer):
    encoded_sentences = vectorizer.transform(sentences)
    len_sentences = [len(sentence.split()) for sentence in sentences]
    feature_names = vectorizer.get_feature_names()

    decoded_sentences = []
    for encoded_sentence in encoded_sentences:
        decoded_sentence = [feature_names[i] for i in encoded_sentence.indices]
        decoded_sentences.append(decoded_sentence)

    return decoded_sentences, len_sentences
```

```python
from catboost import CatBoostClassifier

model = CatBoostClassifier(iterations=3000, learning_rate=0.1, depth=4, l2_leaf_reg=4, loss_function="Logloss",early_stopping_rounds=100)

chunk_size = 25000
total_chunks = len(X_train_final) // chunk_size

start_index = 0
end_index = chunk_size

chunk_data = X_train_final[start_index:end_index]
chunk_targets = y_train.iloc[start_index:end_index]

#print(chunk_data[0], "\n\n")
#print(chunk_targets.values)

model.fit(chunk_data, chunk_targets.values)

for chunk_index in range(1, total_chunks):
    print(chunk_index)
    
    start_index = chunk_index * chunk_size
    end_index = (chunk_index + 1) * chunk_size
    
    chunk_data = X_train_final[start_index:end_index]
    chunk_targets = y_train.iloc[start_index:end_index]
    
    #print(chunk_data[0], "\n\n")
    #print(chunk_targets.values)
    
    model.fit(chunk_data, chunk_targets.values, init_model = model)
```
